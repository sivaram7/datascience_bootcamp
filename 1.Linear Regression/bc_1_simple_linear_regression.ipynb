# Regression iteself

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()  # orginal least square
#fit() will apply specific estimation techniques to OLS in this case to obtain the fit in model
results.summary()
```

```python
plt.scatter(x1,y)
yhat =0.0017*x1 + 0.275
fig= plt.plot(x1,yhat, lw=4, c="green",label="regression line")
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```
